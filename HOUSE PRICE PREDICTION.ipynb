## SPLITTING DATA FOR TRANING AND TESTING

```python
x=df[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade','sqft_living15','sqft_lot15']]
```

```python
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.28,random_state=101)
```

## NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
mx = MinMaxScaler()
x_train = mx.fit_transform(x_train)
x_test = mx.transform(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
x_train_std=std.fit_transform(x_train)
x_test_std=std.transform(x_test)
```

```python
x_train
```

```python
x_train_std
```

```python
x_test_std
```

```python
y_train
```

```python
y_test
```

## MODEL TRANING

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train_std,y_train)
```

```python
y_pred=lr.predict(x_test_std)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
